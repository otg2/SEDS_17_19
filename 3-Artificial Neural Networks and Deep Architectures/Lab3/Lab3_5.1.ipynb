```
import numpy as np




def sgn_single(x):
  return 1 if x > 0 else -1

def sgn(x):
  return np.where(x>0,1,-1)   


x1=np.array([0, 0, 1, 0, 1, 0, 0, 1])
x2=np.array([0, 0, 0, 0, 0, 1, 0, 0])
x3=np.array([0, 1, 1, 0, 0, 1, 0, 1])

x1 = sgn(x1)
x2 = sgn(x2)
x3 = sgn(x3)


dataInput = np.stack((x1,x2,x3))

nNodes = np.shape(dataInput)[1] # Gives us 8 neurons

# Set the coefficient matrix
weights = np.zeros((nNodes,nNodes)) # 8 X 8
nOfDataInputs = np.shape(dataInput)[0] # 3

for i in range(nNodes):
  for j in range(nNodes):
    #if i != j:
      for k in range(nOfDataInputs):
        weights[i,j] += dataInput[k,i]*dataInput[k,j]

print(weights)



def testPatterns(xVec, strVal):
  
  act = np.sum(weights*xVec,axis=1)
  xVec = sgn(act)
  return xVec


activation = np.zeros((nNodes,1))     
  

activation = dataInput[0,:]
a = testPatterns(activation, "x1")
print("x1")
print(np.array_equal(activation,a))

activation = dataInput[1,:]
b = testPatterns(activation, "x2")
print("x3")
print(np.array_equal(activation,b))

activation = dataInput[2,:]
c = testPatterns(activation, "x3")
print("x3")
print(np.array_equal(activation,c))

# Succeeds storing the patterns

```
