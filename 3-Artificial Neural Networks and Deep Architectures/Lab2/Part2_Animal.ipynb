```
# Load data and set up arrays
worksheet = gc.open('Lab2_AnimalData').sheet1
rows = worksheet.get_all_values()
dataArr = pd.DataFrame.from_records(rows)

namesNonTrimmed = dataArr[0]
names = []

for i in range(len(namesNonTrimmed)):
  val = namesNonTrimmed[i]
  if(val != ""):
    names.append([val])

attr = [dataArr[1]]
data = dataArr[2][0]

# 2688
data = data.split(',')

#32 x 84, NxM
#where each row contains the attributes of one animal.
props = np.zeros((np.shape(names)[0], np.shape(attr)[1]))
N = np.shape(props)[0]
M = np.shape(props)[1]


# 32
for i in range(N):
  # 84
  for j in range (M):
    props[i,j] = data[i * M + j]

def checkAnimalAttr(n):
  print("Animal: " + str(names[n]))
  val = np.transpose([props[n], attr[0]])
  print(val)
  
# Antelop is fine 0  
#checkAnimalAttr(0)
#checkAnimalN(1)
def checkAnimalName(n):
  return str(names[n])


#checkAnimalName(5)
```

```
#print(weight1)
from copy import deepcopy

epoch = 20

eta = 0.2
nEta = 0.1
rangeDist = .6

start_eta = deepcopy(eta)
start_nEta = deepcopy(nEta)
start_rangeDist = deepcopy(rangeDist)

final_eta = 0.03
final_nEta = 0.01
final_rangeDist = 0.05

nNodes = 100
weight1 = np.random.rand(nNodes,84)

# Training looppp

for i in range(epoch): # epoch
  
  
  #print("Epoch " + str(i))
  for j in range(N): # N
    
    #pick out the corresponding row from the props matrix.
    animalVal = props[j]
    # find the row of the weight matrix with the shortest distance to this attribute vector
    d = np.linalg.norm(animalVal- weight1,axis = 1)
    # Update winner
    shortestValue = np.min(d)
    shortestIndex = np.argmin(d)
    weight1[shortestIndex,:] += eta * (animalVal - weight1[shortestIndex,:]) 
    # Update neighbours
    dist = np.where(d < shortestValue + rangeDist , 1,0)
    dist[shortestIndex] = 0
    #print(str(np.count_nonzero(dist)) + " neighbours updated")
    weight1 += np.transpose(np.transpose(nEta * (animalVal - weight1)) * dist) 
    
  alphaA = float(i)/float(epoch)
  #learning rates and neighbour update
  eta = start_eta*np.power(final_eta/start_eta,alphaA)
  nEta = start_nEta*np.power(final_nEta/start_nEta,alphaA)

  rangeDist = start_rangeDist*np.power(final_rangeDist/start_rangeDist,alphaA)
    

```

```
# Final loop and sorting
pos = np.zeros([N,1])
for j in range(N):
  animalVal = props[j]
  d = np.linalg.norm(animalVal- weight1,axis = 1) 
  shortestValue = np.min(d)
  shortestIndex = np.argmin(d)
  # Storing the shortest index
  pos[j] = shortestIndex

# Stacking the animalnames onto this and sorting them

stacked = np.stack((pos,names), axis=1)  
stacked = stacked.tolist()

stacked = sorted(stacked, key=lambda stacked_entry: stacked_entry[0])
for i in range(N):
  print(stacked[i])
#print(len(stacked))<

# THIS LOOKS GOOD MANIG
                            
```
